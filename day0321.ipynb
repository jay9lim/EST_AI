```python
# 03customer

cj = pd.read_csv('/content/drive/MyDrive/EST_AI/python_실습파일/week_07/data/bike/customer_join.csv')

# print(cj) # [4192 rows x 18 columns]
# print("- " * 50)
# print()

# print(cj.info()) # column 정보
# print("- " * 50)
# print()

cc =cj[['mean', 'median', 'max', 'min', 'membership_period']]
print(cc) # [4192 rows x 5 columns]

# mean, median, max, min값들과 membership_period숫자차이가나서 Scaler조정  StandardScaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans

sc = StandardScaler()
cc_scaled = sc.fit_transform(cc) # [4192 rows x 5 columns]
df = pd.DataFrame(cc_scaled)
print(df) # [4192 rows x 5 columns]
print("- " * 50)
print()

Kmeans = KMeans(n_clusters=4, random_state=0)
clusters = Kmeans.fit(cc_scaled)
print('clusters = Kmeans.fit(cc_scaled) 훈련')
# print(clusters)
print("- " * 50)
print()

################################################################################
# 매우중요 !!!

# 이 수업 (03/21/2025) 이후 계획
# 분류, 회귀, 클러스터링 (오늘끝마칠) -> 다음주 공공데이터수업 프로젝트 진행
# 분류, 회귀, 클러스트링 (오늘끝마칠) -> 공공데이터 프로젝행 진행후 웹크롤링 수업 + 프로젝트 진행
# (공공데이터수업+1차프로젝트진행) + (웹크롤링수업+2차프로젝트진행) + (다시 ML진행 + 딥러닝)
# (3차)최종프로젝트 -> (ML진행 + 딥러닝) + AI 연결 +(오전수업진행)
################################################################################

# cc =cj[['mean', 'median', 'max', 'min', 'membership_period', 열컬럼하나추가]]
# print(clusters.labels_)
cc['cluster'] = clusters.labels_
print(cc) # [4192 rows x 19 columns]
print()
print("- " * 50)

print(cc['cluster'].unique()) # [2 1 0 3]
print("- " * 50)
print()

# 이름변경 ex) mean -> 월평균값
cc.rename(columns = {'mean':'월평균값', 'median':'월중앙값', 'max':'월최대값',
                     'min':'월최소값', 'membership_period':'회원기간'}, inplace=True)
print(cc) #[4192 rows x 6 columns]
print("- " * 50)
print()
print('***** cluster항목으로 그룹화 갯수 mean *****')
print(cc.groupby('cluster').mean())
print("- " * 50)
print()

#시각화작업 4행*5열 sns.scatterplot()
from sklearn.decomposition import PCA

X = cc_scaled.copy()
pca = PCA(n_components=2)
pca.fit(X)

x_pca = pca.transform(X)
pca_df = pd.DataFrame(x_pca)
pca_df['cluster'] = cc['cluster']
print(pca_df) #cluster명시해서
print("- " * 50)
print()

# 아래내용 권장
sns.scatterplot(data=pca_df, x=0, y=1, hue='cluster')
plt.show()
print("- " * 50)
print()

# 비권장
# for i in (pca_df ["cluster"].unique()):
#     tmp = pca_df.loc[pca_df["cluster"] == i]
#     plt.scatter(tmp[0], tmp[1])
#     plt.legend(pca_df["cluster"].unique())

# plt.show()
# print("- " * 50)
# print()


# cluster별, is_deleted컬럼 그룹화의 갯수 count()
cc_join = pd.concat([cc, cj], axis=1) #cc & cj 연결
# print(cc_join)
new_df = cc_join.groupby(['cluster','is_deleted'],
                          as_index=False).count()[['cluster','is_deleted','customer_id']]
print(new_df)
print("- " * 50)
print()

delete0 = (new_df['is_deleted'] == 0) # 0 현재 지속회원 유지
delete1 = (new_df['is_deleted'] == 1) # 1 현재 지속회원 유지


for i in range(0,4):
    temp = (new_df['cluster'] == i)
    print('Cluster ' + str(i) + '의 탈퇴회원 비율은' + str( round( (new_df.loc[(temp&delete1), 'customer_id'].sum() / new_df.loc[(temp), 'customer_id'].sum())*100 ,2)) + '%' )
    #힌트 delete1변수사용 ,  new_df , sum()
    # str( round( (new_df.loc[(temp&delete1), 'customer_id'].sum() / new_df.loc[(temp), 'customer_id'].sum())*100 ,2))
    print('Cluster ' + str(i) + '의 지속회원 비율은' + str( round( (new_df.loc[(temp&delete0), 'customer_id'].sum() / new_df.loc[(temp), 'customer_id'].sum())*100 ,2)) + '%' )
    print()
```
